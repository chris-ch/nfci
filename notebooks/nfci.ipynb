```python
import pandas
```

```python
nfci = pandas.read_csv('https://www.chicagofed.org/-/media/publications/nfci/nfci-data-series-csv.csv?sc_lang=en')
nfci.set_index('Friday_of_Week', inplace=True)
```

```python
nfci.columns
```

```python
import matplotlib
from cycler import cycler
matplotlib.style.use('ggplot')
matplotlib.rcParams['lines.linewidth'] = 3
matplotlib.rcParams['axes.prop_cycle'] = cycler('color', ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'])
```

```python
nfci['NFCI'].plot(figsize=(16, 4), grid=True)
```

```python
nfci['ANFCI'].plot(figsize=(16, 4), grid=True)
```

```python
nfci[['Risk', 'Credit', 'Leverage', 'Nonfinancial_Leverage']].plot(figsize=(16, 4), grid=True)
```
